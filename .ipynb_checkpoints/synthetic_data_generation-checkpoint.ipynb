```python
# Synthetic Data Generation with Gaussian Copula

import os
import sys
import time
import logging
import warnings
from typing import Optional, Dict, Any, List, Tuple

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, HTML, Markdown
from sdv.metadata import SingleTableMetadata
from sdv.single_table import GaussianCopulaSynthesizer
from sdv.evaluation.single_table import evaluate_quality
```

```python


# Configure matplotlib and seaborn for better visualizations
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.2)

# Set up basic logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)

# Suppress certain warnings that might clutter the notebook
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=DeprecationWarning)

# Display function for better notebook output formatting
def display_header(text):
    display(HTML(f"<h3>{text}</h3>"))
    
def display_subheader(text):
    display(HTML(f"<h4>{text}</h4>"))
```

```python
def load_data(filepath: str) -> pd.DataFrame:
    """
    Load data from CSV file with error handling.
    
    Args:
        filepath: Path to the CSV file
        
    Returns:
        Loaded DataFrame
    """
    display_header("Data Loading")
    
    try:
        if not os.path.exists(filepath):
            raise FileNotFoundError(f"File not found: {filepath}")
            
        print(f"Loading data from {filepath}...")
        df = pd.read_csv(filepath)
        print(f"Successfully loaded data with shape {df.shape}")
        return df
    except Exception as e:
        print(f"Error loading data: {str(e)}")
        raise
```

```python
def explore_data(df: pd.DataFrame) -> None:
    """
    Perform exploratory data analysis on the input DataFrame with nicely aligned plots.
    
    Args:
        df: Input DataFrame
    """
    display_header("Exploratory Data Analysis")
    
    # Display basic information
    display_subheader("Data Overview")
    print(f"Dataset shape: {df.shape}")
    display(df.head())
    
    # Display data types and memory usage
    display_subheader("Data Types and Memory Usage")
    dtype_info = pd.DataFrame({
        'Data Type': df.dtypes,
        'Memory Usage (MB)': df.memory_usage(deep=True) / 1024 / 1024
    })
    display(dtype_info)
    
    # Display summary statistics
    display_subheader("Summary Statistics")
    display(df.describe().T.style.background_gradient(cmap='Blues', axis=1))
    
    # Count missing values
    display_subheader("Missing Values")
    missing = df.isnull().sum()
    missing_percent = (missing / len(df)) * 100
    missing_info = pd.DataFrame({
        'Missing Values': missing,
        'Percentage (%)': missing_percent.round(2)
    })
    
    if missing.sum() > 0:
        # Display missing values info with styling
        display(missing_info[missing_info['Missing Values'] > 0]
                .sort_values('Missing Values', ascending=False)
                .style.background_gradient(cmap='Reds'))
        
        # Visualize missing values
        plt.figure(figsize=(12, 6))
        cols_with_missing = missing_info[missing_info['Missing Values'] > 0].index
        if len(cols_with_missing) > 0:
            sns.heatmap(df[cols_with_missing].isnull(), 
                      cmap='viridis', 
                      yticklabels=False, 
                      cbar_kws={'label': 'Missing Values'})
            plt.title('Missing Value Patterns')
            plt.tight_layout()
            plt.show()
    else:
        print("No missing values in the dataset.")
    
    # Visualize distributions of numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    
    if len(numerical_cols) > 0:
        display_subheader("Distribution of Numerical Features")
        
        # Calculate optimal grid layout based on number of columns
        sample_cols = numerical_cols[:min(12, len(numerical_cols))]
        num_cols = len(sample_cols)
        num_rows = (num_cols + 2) // 3  # 3 plots per row, rounded up
        
        # Create subplots grid with better layout
        fig = plt.figure(figsize=(18, num_rows * 4))
        grid = plt.GridSpec(num_rows, 3, figure=fig, hspace=0.4, wspace=0.3)
        
        for i, col in enumerate(sample_cols):
            row, col_pos = divmod(i, 3)
            ax = fig.add_subplot(grid[row, col_pos])
            
            # Create a more informative plot combining histogram and boxplot
            sns.histplot(df[col].dropna(), kde=True, ax=ax, color='skyblue', alpha=0.7)
            
            # Add mean and median lines
            mean_val = df[col].mean()
            median_val = df[col].median()
            ax.axvline(mean_val, color='red', linestyle='--', label=f'Mean: {mean_val:.2f}')
            ax.axvline(median_val, color='green', linestyle=':', label=f'Median: {median_val:.2f}')
            
            # Add descriptive statistics as text
            stats_text = (f"Std: {df[col].std():.2f}\n"
                        f"Min: {df[col].min():.2f}\n"
                        f"Max: {df[col].max():.2f}")
            props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
            ax.text(0.05, 0.95, stats_text, transform=ax.transAxes, fontsize=9,
                  verticalalignment='top', bbox=props)
            
            ax.set_title(f'Distribution of {col}')
            ax.legend(fontsize='small')
            
        plt.tight_layout()
        plt.show()
        
        # Display correlation matrix with improved visualization
        if len(numerical_cols) > 1:
            display_subheader("Correlation Matrix")
            
            # Create correlation matrix
            corr = df[numerical_cols].corr().round(2)
            
            # Generate a mask for the upper triangle
            mask = np.triu(np.ones_like(corr, dtype=bool))
            
            # Set up the matplotlib figure
            plt.figure(figsize=(12, 10))
            
            # Draw the heatmap with the mask
            sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', 
                      fmt=".2f", linewidths=0.5, vmin=-1, vmax=1, 
                      annot_kws={"size": 10})
            plt.title('Correlation Matrix (Lower Triangle Only)', fontsize=14)
            
            # Rotate x-axis labels for better readability
            plt.xticks(rotation=45, ha='right', fontsize=10)
            plt.yticks(fontsize=10)
            
            # Adjust layout to prevent clipping
            plt.tight_layout()
            plt.show()
            
            # If there are many numerical columns, show the highest correlations separately
            if len(numerical_cols) > 5:
                # Extract the highest absolute correlations (excluding self-correlations)
                corr_unstack = corr.unstack()
                corr_abs = corr_unstack.apply(abs)
                # Remove self-correlations (which are always 1.0)
                corr_abs = corr_abs[corr_abs < 1.0]
                highest_corrs = corr_abs.sort_values(ascending=False).head(15)
                
                display_subheader("Top Correlations")
                for (col1, col2), corr_val in highest_corrs.items():
                    actual_val = corr.loc[col1, col2]
                    print(f"{col1} — {col2}: {actual_val:.2f}")
                
                # Plot the top correlations as scatter plots
                pairs_to_plot = [(idx[0], idx[1]) for idx in highest_corrs.index][:6]  # Take top 6 pairs
                
                if pairs_to_plot:
                    fig = plt.figure(figsize=(18, 12))
                    grid = plt.GridSpec(2, 3, figure=fig, hspace=0.3, wspace=0.3)
                    
                    for i, (col1, col2) in enumerate(pairs_to_plot):
                        row, col_pos = divmod(i, 3)
                        ax = fig.add_subplot(grid[row, col_pos])
                        
                        # Create scatter plot with trend line
                        sns.regplot(x=df[col1], y=df[col2], ax=ax, scatter_kws={'alpha':0.5})
                        r_value = df[col1].corr(df[col2])
                        ax.set_title(f'{col1} vs {col2} (r = {r_value:.2f})')
                        
                    plt.tight_layout()
                    plt.show()
    
    # Visualize categorical columns
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    
    if len(categorical_cols) > 0:
        display_subheader("Categorical Features")
        
        # Calculate optimal grid layout for categorical plots
        sample_cat_cols = categorical_cols[:min(6, len(categorical_cols))]
        num_cat_cols = len(sample_cat_cols)
        num_cat_rows = (num_cat_cols + 1) // 2  # 2 plots per row, rounded up
        
        fig = plt.figure(figsize=(18, num_cat_rows * 5))
        grid = plt.GridSpec(num_cat_rows, 2, figure=fig, hspace=0.4, wspace=0.2)
        
        for i, col in enumerate(sample_cat_cols):
            row, col_pos = divmod(i, 2)
            ax = fig.add_subplot(grid[row, col_pos])
            
            # Get top categories (handle large number of categories)
            value_counts = df[col].value_counts().sort_values(ascending=False)
            top_n = min(10, len(value_counts))  # Take top 10 or fewer if there are less
            
            if len(value_counts) > top_n:
                # If there are too many categories, group the smallest ones as "Other"
                top_values = value_counts.head(top_n-1)
                other_count = value_counts.iloc[top_n-1:].sum()
                plot_data = pd.concat([top_values, pd.Series({'Other': other_count})])
            else:
                plot_data = value_counts
            
            # Create horizontal bar chart for better readability of category names
            sns.barplot(x=plot_data.values, y=plot_data.index, ax=ax, palette='viridis')
            ax.set_title(f'Distribution of {col} (Total: {len(value_counts)} unique values)')
            ax.set_xlabel('Count')
            
            # Add percentages to the bars
            total = plot_data.sum()
            for i, v in enumerate(plot_data.values):
                percentage = v / total * 100
                ax.text(v + 0.1, i, f'{percentage:.1f}%', va='center')
                
        plt.tight_layout()
        plt.show()
        
        # For columns with few unique values, show their relationship with a key numerical feature
        if numerical_cols and len(categorical_cols) > 0:
            display_subheader("Categorical-Numerical Relationships")
            
            # Choose a numerical column with highest variance for analysis
            numerical_variances = df[numerical_cols].var()
            target_numerical = numerical_variances.idxmax()
            
            # Create box plots for categorical features vs. target numerical
            sample_cat_for_box = [col for col in categorical_cols 
                                if df[col].nunique() <= 15][:4]  # Max 4 columns with ≤15 unique values
            
            if sample_cat_for_box:
                fig = plt.figure(figsize=(18, 5 * len(sample_cat_for_box)))
                
                for i, cat_col in enumerate(sample_cat_for_box):
                    ax = fig.add_subplot(len(sample_cat_for_box), 1, i+1)
                    
                    # Order by median of numerical value
                    order = df.groupby(cat_col)[target_numerical].median().sort_values().index
                    
                    # Create box plot
                    sns.boxplot(x=cat_col, y=target_numerical, data=df, ax=ax, 
                              order=order, palette='Set3')
                    ax.set_title(f'{cat_col} vs {target_numerical}')
                    ax.set_xlabel(cat_col)
                    ax.set_ylabel(target_numerical)
                    plt.xticks(rotation=45, ha='right')
                    
                plt.tight_layout()
                plt.show()
```

```python
def validate_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Validate and clean the input data.
    
    Args:
        df: Input DataFrame
        
    Returns:
        Cleaned DataFrame
    """
    display_header("Data Validation")
    
    # Report on missing values
    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print(f"Missing values detected:")
        display(missing_values[missing_values > 0])
    else:
        print("No missing values detected.")
    
    # Check for duplicates
    duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicates}")
    
    # Check for potential outliers in numerical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
    if len(numerical_cols) > 0:
        display_subheader("Checking for Outliers")
        for col in numerical_cols:
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            outliers = df[(df[col] < Q1 - 1.5 * IQR) | (df[col] > Q3 + 1.5 * IQR)][col].count()
            if outliers > 0:
                print(f"Potential outliers in {col}: {outliers} ({outliers/len(df)*100:.2f}%)")
    
    # Check for column data types
    display_subheader("Data Types")
    display(df.dtypes)
    
    print("\nData validation complete. You can now proceed with data cleaning if needed.")
    
    return df
```

```python
def generate_metadata(df: pd.DataFrame) -> SingleTableMetadata:
    """
    Generate metadata for the DataFrame.
    
    Args:
        df: Input DataFrame
        
    Returns:
        SDV SingleTableMetadata object
    """
    display_header("Metadata Generation")
    
    print("Generating metadata from the DataFrame...")
    metadata = SingleTableMetadata()
    metadata.detect_from_dataframe(df)
    
    # Display detected column types
    column_types = {column: props['sdtype'] for column, props in metadata.columns.items()}
    column_types_df = pd.DataFrame.from_dict(column_types, orient='index', columns=['Detected Type'])
    
    display(column_types_df)
    print("Metadata generation complete.")
    
    return metadata

```

```python
def generate_synthetic_data(
    df: pd.DataFrame,
    metadata: Optional[SingleTableMetadata] = None,
    num_samples: int = 10000,
    enforce_min_max_values: bool = True,
    enforce_rounding: bool = True,
    model_params: Optional[Dict[str, Any]] = None
) -> pd.DataFrame:
    """
    Generate synthetic data using GaussianCopulaSynthesizer.
    
    Args:
        df: Input DataFrame
        metadata: Optional pre-generated metadata
        num_samples: Number of synthetic samples to generate
        enforce_min_max_values: Whether to enforce min/max values
        enforce_rounding: Whether to enforce rounding in numerical columns
        model_params: Additional parameters for the GaussianCopulaSynthesizer
        
    Returns:
        DataFrame of synthetic data
    """
    display_header("Synthetic Data Generation")
    
    # Generate metadata if not provided
    if metadata is None:
        metadata = generate_metadata(df)
    
    # Set default model parameters if not provided
    if model_params is None:
        model_params = {}
    
    # Initialize the synthesizer
    print("Initializing Gaussian Copula synthesizer...")
    synthesizer = GaussianCopulaSynthesizer(
        metadata=metadata,
        enforce_min_max_values=enforce_min_max_values,
        enforce_rounding=enforce_rounding,
        **model_params
    )
    
    # Fit the model
    print(f"Fitting the synthesizer on {len(df)} rows of data...")
    start_time = time.time()
    synthesizer.fit(df)
    fit_time = time.time() - start_time
    print(f"Fitting completed in {fit_time:.2f} seconds")
    
    # Generate synthetic data
    print(f"Generating {num_samples} synthetic samples...")
    start_time = time.time()
    synthetic_data = synthesizer.sample(num_samples)
    sample_time = time.time() - start_time
    print(f"Sampling completed in {sample_time:.2f} seconds")
    
    print(f"Generated synthetic data with shape: {synthetic_data.shape}")
    display(synthetic_data.head())
    
    return synthetic_data
```

```python
def compare_distributions(real_data: pd.DataFrame, synthetic_data: pd.DataFrame, columns: Optional[List[str]] = None) -> None:
    """
    Compare distributions between real and synthetic data.
    
    Args:
        real_data: Original DataFrame
        synthetic_data: Synthetic DataFrame
        columns: List of columns to compare (if None, selects a sample of numerical columns)
    """
    display_header("Comparing Real vs Synthetic Distributions")
    
    # Select columns to compare if not specified
    if columns is None:
        numerical_cols = real_data.select_dtypes(include=['int64', 'float64']).columns.tolist()
        columns = numerical_cols[:min(5, len(numerical_cols))]
    
    # Compare distributions for selected columns
    for col in columns:
        plt.figure(figsize=(12, 6))
        
        # Plot histograms
        plt.subplot(1, 2, 1)
        sns.histplot(real_data[col].dropna(), kde=True, color='blue', label='Real', alpha=0.6)
        sns.histplot(synthetic_data[col].dropna(), kde=True, color='red', label='Synthetic', alpha=0.6)
        plt.title(f'Distribution of {col}')
        plt.legend()
        
        # Plot box plots
        plt.subplot(1, 2, 2)
        plt.boxplot([real_data[col].dropna(), synthetic_data[col].dropna()], 
                  labels=['Real', 'Synthetic'])
        plt.title(f'Box Plot of {col}')
        
        plt.tight_layout()
        plt.show()
    
    # Compare summary statistics
    display_subheader("Summary Statistics Comparison")
    real_stats = real_data[columns].describe()
    synthetic_stats = synthetic_data[columns].describe()
    
    for col in columns:
        display(Markdown(f"**Column: {col}**"))
        comparison = pd.DataFrame({
            'Real': real_stats[col],
            'Synthetic': synthetic_stats[col],
            'Difference': real_stats[col] - synthetic_stats[col],
            'Diff %': ((real_stats[col] - synthetic_stats[col]) / real_stats[col] * 100).round(2)
        })
        display(comparison)
```

```python
def evaluate_synthetic_data(real_data: pd.DataFrame, synthetic_data: pd.DataFrame) -> dict:
    """
    Evaluate the quality of synthetic data using SDV's evaluation metrics.
    
    Args:
        real_data: Original DataFrame
        synthetic_data: Synthetic DataFrame
        
    Returns:
        Evaluation metrics
    """
    display_header("Synthetic Data Quality Evaluation")
    
    try:
        # Use SDV's evaluation module
        print("Evaluating synthetic data quality...")
        quality_report = evaluate_quality(
            real_data=real_data,
            synthetic_data=synthetic_data,
            metadata=SingleTableMetadata()
        )
        
        # Display overall quality score
        score = quality_report.get_score()
        print(f"Overall Quality Score: {score:.4f} (0-1 scale, higher is better)")
        
        # Display properties scores if available
        try:
            properties = quality_report.get_properties()
            properties_df = pd.DataFrame(properties)
            display(properties_df)
        except:
            print("Detailed properties not available.")
        
        # Display visualization if available
        try:
            fig = quality_report.get_visualization()
            plt.figure(figsize=(10, 8))
            plt.tight_layout()
            plt.show()
        except:
            print("Quality visualization not available.")
            
        return quality_report
    except Exception as e:
        print(f"Could not generate quality report: {str(e)}")
        return {"error": str(e)}
```

```python
def save_data(df: pd.DataFrame, output_path: str) -> None:
    """
    Save DataFrame to CSV.
    
    Args:
        df: DataFrame to save
        output_path: Path for the output file
    """
    display_header("Save Synthetic Data")
    
    try:
        # Create directory if it doesn't exist
        os.makedirs(os.path.dirname(output_path), exist_ok=True)
        
        print(f"Saving synthetic data to {output_path}")
        df.to_csv(output_path, index=False)
        print(f"Successfully saved data with shape {df.shape}")
    except Exception as e:
        print(f"Error saving data: {str(e)}")
        raise

```
